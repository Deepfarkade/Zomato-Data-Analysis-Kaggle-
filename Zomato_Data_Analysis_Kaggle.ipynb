# Exploratory Data Analysis of Zomato Bangalore Restaurants

```python
df = df.drop(['url','address','phone','dish_liked','menu_item','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

```python
# Dropping the duplicates rows
df.drop_duplicates(inplace = True)
df.shape
```

```python
# Cleaning the rate column
df['rate'].unique()
```

```python
# Removing "NEEW","-","/5" from rate column. for tht we will create a function

def handlerate(value):
    if (value =="NEW" or value == "-"):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
# Filling the null values in the rate column
# checking the how many are the null values

df.rate.isnull().sum()
```

```python
# we are filling this null values with the mean of the rate column

df['rate'].fillna(df['rate'].mean(),inplace = True)
df['rate'].isnull().sum()

# Now we can clearly see that all the null values are filled so ther is no null values
```

```python
## There are not too much null values remaining in the data set so we decided to remove or drop the null values

df.dropna(inplace = True)
df.head()
```

```python
# So now there are too many columns names in the data set which are lenghty so will rename that columns

df.rename(columns = {'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
# So both the columns have basically same areas an dthe same values so will delete one colummn i.e city walla

df = df.drop(['listed_in(city)'],axis = 1)
df.head()
```

```python
df['Cost2plates'].unique()
```

```python
df['Cost2plates'].dtype
```

```python
# So we can clearly see that there are ',' in the values of cost2plates so we have to remove this
# because tht's why it is showing us in object and we want float or intiger

def handlecomma (value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma) 
df['Cost2plates'].unique()
```

```python
df.head()
```

```python
# Cleaining the Rest type column

df['rest_type'].value_counts()
```

```python
# We arejust storing the bove list in the rest_type so as to make the cluster i.e the restaurants having less that 1000 rest_type
# so we will cluster them all in one and make one another column for them which will contains all the rest types less than 1000.


rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_type_lessthan1000 = rest_types[rest_types<1000]
rest_type_lessthan1000
```

```python
# Now making the function for which if the rest type which we calculated above is less than 1000
# so that rest type will be named as others in the main rest_type column

def handle_rest_type(value):
    if value in rest_type_lessthan1000:
        return "Others"
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
# Now cleaning the Location Column

df['location'].value_counts()
```

```python
# Same as we did above (Makeing the cluster)

location = df['location'].value_counts(ascending = False)
```

```python
location
```

```python
location_leassthan300 = location[location<300]

location_leassthan300
```

```python
def handle_location(value):
    if value in location_leassthan300:
        return "Others"
    else:
        return value
    
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
## Now cleaning the Cusins Column

df['cuisines'].value_counts()
```

```python
# Same as we did above that will make the clusters

cuisines = df['cuisines'].value_counts(ascending = False)

cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if value in cuisines_lessthan100:
        return "Others"
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

# VISUALIZATION OF THE DATA

```python
# count plot of various locations

plt.figure(figsize = (16,10))
ax = sns.countplot(df['location'])
plt.xticks(rotation = 90)
```

```python
# Now visualizing that how many restaurants have the online order facility and how many don't have the online Ordering Facility

plt.figure(figsize = (6,6))
sns.countplot(df['online_order'],palette = 'inferno')
```

```python
# Now visualizing that how many restaurants have the book table facility and how many don't have the book table Facility

plt.figure(figsize = (6,6))
sns.countplot(df['book_table'],palette = 'inferno')
```

```python
# Now visualizing the online order vs rate (ratings from - 0 t0 5)

plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order',y = 'rate',data = df)
```

```python
# Now analyzing the Book table vs the rate

plt.figure(figsize = (6,6))
sns.boxplot(x = 'book_table', y = 'rate',data = df)
```

```python
# Visualizing the online Order facility location wise

df1 = df.groupby(['location','online_order'])['name'].count()
df1.to_csv('location_online.csv')
df1 = pd.read_csv('location_online.csv')
df1 = pd.pivot_table(df1,values = None,index = ['location'],columns = ['online_order'],fill_value = 0, aggfunc = np.sum)
df1
```

```python
df1.plot(kind='bar', figsize = (15,8))
```

```python
# Visualizing the online book_table location wise

df2 = df.groupby(['location','book_table'])['name'].count()
df2.to_csv('location_book_table.csv')
df2 = pd.read_csv('location_book_table.csv')
df2 = pd.pivot_table(df2,values = None,index = ['location'],columns = ['book_table'],fill_value = 0, aggfunc = np.sum)
df2
```

```python
df2.plot(kind='bar', figsize = (15,8))
```

```python
# Visualizing types of restaurants vs rate

plt.figure(figsize = (14,8))
sns.boxplot(x = 'Type', y = 'rate', data = df)
```

```python
# Grouping the types of restaurants location wise

df3 = df.groupby(['location','Type'])['name'].count()
df3.to_csv('location_Type.csv')
df3 = pd.read_csv('location_Type.csv')
df3 = pd.pivot_table(df3,values = None,index = ['location'],columns = ['Type'],fill_value = 0, aggfunc = np.sum)
df3
```

```python
df3.plot(kind='bar', figsize = (20,8))
```

```python
# Now visualizing that in which location no of people are voting the most (location vs votes)

df4 = df[['location','votes']]
df4.drop_duplicates()
df5 = df4.groupby(['location'])['votes'].sum()
df5 = df5.to_frame()
df5 = df5.sort_values('votes',ascending=False)
df5.head()
```

```python
plt.figure(figsize = (15,8))
sns.barplot(df5.index, df5['votes'])
plt.xticks(rotation = 90)

# So in this visualization we can clearly see that people in the koramangala 5th block are intrested in voting and same for indiranagar and etc
```
